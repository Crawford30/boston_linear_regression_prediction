```python
from sklearn.datasets import load_boston
```

```python
boston_dataset= load_boston()
```

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']
target = df_boston['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = df_boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

concatenating the using the numpy library

```python
X = pd.DataFrame(np.c_[df_boston['LSTAT'], df_boston['RM']], columns = ['LSTAT','RM'])
Y = df_boston['MEDV']
```

splitting the data into both training and tests data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=101)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Training and testing the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

model evaluation using the RMSE

Model Evaluation for Training set


```python
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")

print('RMSE is  {}'.format(rmse))
print('R2 is {}'.format(r2))
print("\n")
```

Model evaluation for Test set

```python
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
